````python
import json 


# 示例数据：多个 external_params 和 room_layout 的组合 

data_pairs = [ 

    { 

        "external_params": { 

            "边界": [0, 0, 9300, 11100], 

            "黑体1": [0, 9900, 4200, 1200], 

            "南采光": [0, -1200, 9300, 1200], 

            "北采光": [0, 11100, 9300, 1200], 

            "东采光": [9300, 0, 1200, 11100], 

            "主入口": [300, 9900, 1200, 1200] 

        }, 

        "room_layout": { 

            "客厅": [0, 1800, 3600, 2700], 

            "卧室1": [3600, 1800, 2400, 2700], 

            "卧室2": [6000, 0, 3300, 3600], 

            "采光1": [4200, 8400, 5100, 2700], 

            "卧室3": [6600, 5400, 2700, 3000], 

            "卫生间": [3300, 6000, 2100, 2400], 

            "厨房": [1800, 7800, 2400, 2100], 

            "采光2": [0, 0, 6000, 1800], 

            "采光3": [7200, 3600, 2100, 1800], 

            "餐厅": [1800, 6000, 1500, 1800] 

        }, 

        "image_path": "xxx/x1.jpg" 

    }, 

    { 

        "external_params": { 

            "边界": [0, 0, 10000, 12000], 

            "黑体1": [0, 10800, 4500, 1200], 

            "南采光": [0, -1200, 10000, 1200], 

            "北采光": [0, 12000, 10000, 1200], 

            "东采光": [10000, 0, 1200, 12000], 

            "主入口": [400, 10800, 1200, 1200] 

        }, 

        "room_layout": { 

            "客厅": [0, 2000, 4000, 3000], 

            "卧室1": [4000, 2000, 2500, 3000], 

            "卧室2": [6500, 0, 3500, 4000], 

            "采光1": [4500, 9000, 5500, 3000], 

            "卧室3": [7000, 6000, 3000, 3500], 

            "卫生间": [3500, 6500, 2500, 2500], 

            "厨房": [2000, 8500, 2500, 2500], 

            "采光2": [0, 0, 6500, 2000], 

            "采光3": [7500, 4000, 2500, 2000], 

            "餐厅": [2000, 6500, 1500, 2000] 

        }, 

        "image_path": "xxx/x2.jpg" 

    } 

] 


# 构建 ShareGPT 格式的 JSON 数据 

sharegpt_data = [] 


for pair in data_pairs: 

    external_params = pair["external_params"] 

    room_layout = pair["room_layout"] 

    image_path = pair["image_path"] 


    # 将数据转换为 JSON 格式字符串 

    external_json = json.dumps(external_params, ensure_ascii=False) 

    rooms_list = json.dumps(list(room_layout.keys()), ensure_ascii=False) 

    layout_json = json.dumps(room_layout, ensure_ascii=False) 


    # 构建单个 ShareGPT 消息 

    sharegpt_entry = { 

        "messages": [ 

            { 

                "content": f"<image>请根据这张图片中的外部环境帮我设计内部房间排布。外部环境的参数为：\n```json\n{external_json}\n```\n待生成的房间名称为：\n```json\n{rooms_list}```", 

                "role": "user" 

            }, 

            { 

                "content": f"```json\n{layout_json}\n```", 

                "role": "assistant" 

            } 

        ], 

        "images": [image_path] 

    } 


    # 添加到最终结果中 

    sharegpt_data.append(sharegpt_entry) 


# 输出格式化后的 JSON 

print(json.dumps(sharegpt_data, ensure_ascii=False, indent=2)) 
````
